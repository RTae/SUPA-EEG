```python
import torch
from torchview import draw_graph
from src.encoders.eegnet_encoder import EEGNetEncoder
from src.models.supaeeg import SUPAEEG, SubjectAwareRouter
```

```python
model = EEGNetEncoder(n_channels=17, n_timepoints=100, feature_dim=1024)
graph = draw_graph(
    model,
    input_size=(1, 17, 100),
    graph_name="EEGNetEncoder",
    expand_nested=True,
    save_graph=True,
    filename="eegnet_encoder",
    directory="reports",
)
```

```python
from pathlib import Path
from IPython.display import Image, display

saved = Path("reports") / "eegnet_encoder.png"
print(f"Rendered file: {saved.resolve()}")

if saved.exists():
    display(Image(filename=str(saved)))
else:
    print("PNG not found yet. The graph object is available in `graph.visual_graph`.")
```

```python
full_model = SUPAEEG()
full_graph = draw_graph(
    full_model,
    input_size=[(1, 17, 100), (1, 5, 3200), (1,)],
    dtypes=None,
    graph_name="SUPAEEG",
    expand_nested=True,
    save_graph=True,
    filename="supaeeg_model",
    directory="reports",
)
```

```python
full_saved = Path("reports") / "supaeeg_model.png"
print(f"Rendered file: {full_saved.resolve()}")

if full_saved.exists():
    display(Image(filename=str(full_saved)))
else:
    print("PNG not found yet. The graph object is available in `full_graph.visual_graph`.")
```

```python
router = SubjectAwareRouter(n_subjects=10, n_layers=5)
router.train()

subject_ids = torch.tensor([0], dtype=torch.long)
router_graph = draw_graph(
    router,
    input_data=[subject_ids],
    graph_name="SubjectAwareRouter",
    expand_nested=True,
    save_graph=True,
    filename="subject_aware_router",
    directory="reports",
)
```

```python
router_saved = Path("reports") / "subject_aware_router.png"
print(f"Rendered file: {router_saved.resolve()}")

if router_saved.exists():
    display(Image(filename=str(router_saved)))
else:
    print("PNG not found yet. The graph object is available in `router_graph.visual_graph`.")
```
